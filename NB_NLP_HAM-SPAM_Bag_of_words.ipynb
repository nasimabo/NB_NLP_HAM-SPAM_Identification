```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('spam.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.Message,df.Category,test_size = 0.3)
```

```python
x_train
```

```python
x_train.values
```

```python
x_test
```

# x_train bag_of_words

```python
from sklearn.feature_extraction.text import CountVectorizer
fe = CountVectorizer()
x_train_cv = fe.fit_transform(x_train.values)
```

```python
x_train_cv.toarray()
```

```python
x_train_cv.shape
```

```python
fe.get_feature_names_out()
```

```python
pd.DataFrame(data = x_train_cv.toarray() , columns = fe.get_feature_names_out())
```

# naive bayes model

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_cv,y_train)
```

# x_test bag_of_words

```python
x_test_cv = fe.transform(x_test)
```

```python
x_test_cv.toarray()
```

```python
x_test_cv.shape
```

```python
fe.get_feature_names_out(x_test)
```

```python
pd.DataFrame(data = x_test_cv.toarray(),columns = fe.get_feature_names_out())
```

```python
model.score(x_test_cv,y_test)
```
